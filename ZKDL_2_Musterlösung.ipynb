# Filtern und Auswählen

```python
# an dieser sptelle können wir filtern
publications = [item for item in unique_raw_entries if ((item['type'] == 'Publication') and int(item["year"])>2003)]
```

```python
# Auswahl der Informationen, die ausgewählt werden sollen:
wanted_attributes = ['isbn',
 'editor',
 'year',
 'publisher',
 'journal',
 'abstract',
 'tags',
 'refid',
 'doi',
 'number',
 'label',
 'author',
 'interHash',
 'count',
 'editors',
 'date',
 'id',
 'url',
 'type',
 'note',
 'volume',
 'changeDate',
 'pages',
 'description']
```

```python
edited_items = []

# für jedes Element im eingelesenen JSON file... (über data.keys() lässt sich leicht herausfinden, wo die relevanten Daten stecken):

for item in unique_raw_entries:

    # erstelle ein neues leeres dictionary für die Daten des Elements

    item_export = {}

    # für jeden gewünschten Eintrag

# An dieser Stelle (bevor wir den Eintrag übernehmen) können wir Einträge aussortieren. Alternativ können auch erst einmal alle Einträge ins gewünsche Format gebracht werden (und auf die benötigten Informationen reduziert werden) und dann erst beim Schreiben des XML-Files aussortiert werden.


    # hier können Filterkriterien hinzugefügt werden (hier z.B. erschienen nach 2009 und mit weniger als 4 Autoren):
    if "year" in item.keys() and "author" in item.keys():
        if int(item["year"])>1009  and len(item["author"])<4:
            for entry in wanted_attributes:

                # ... überprüfe, ob es einen Eintrag gibt (um Fehler zu vermeiden)

                if entry in item.keys():

                    # füge den Eintrag den Daten zum Element hinzu

                    item_export[entry] = item[entry]

            # am Ende: füge den vollständigen (gefilterten) Eintrag der zu exportierenden Liste hinzu

            edited_items.append(item_export)    
```

```python
from xml.sax.saxutils import escape
```

```python
print(escape(a))
```

```python
with open("output.xml", "w") as xml:
    
    print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>", file=xml)
    print("<publications>", file=xml)
    
    for item in edited_items:
        print("\t<publication>", file=xml)
        
        for key in item.keys():

            print("\t\t"+"<"+str(key)+">"+escape(str(item[key]))+"</"+str(key)+">", file=xml)
                
        print("\t</publication>", file=xml)
                
    print("</publications>\n", file=xml)
```

```python
with open("outfile.xml", "w") as xml:
    print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>", file=xml)
    
    print("<add>", file=xml)
    for item in edited_items:
        
            print("\t<doc>", file=xml)
            for key in item.keys():   
                
                print("\t\t"+"<"+escape(str(key))+">"+escape(str(item[key]))+"</"+escape(str(key))+">", file=xml)
            print("\t</doc>", file=xml)
            
    print("</add>\n", file=xml)
```
