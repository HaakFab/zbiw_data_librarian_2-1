# 1: Writing XML

```python
cars = [
    {
  "brand": "Peugeot",
"l": ["Angela Merkel", 2, 3, 4],
  "model": "206",
  "year": 1985
},{
  "brand": "VW",
        "l": ["Peter Müller", 2, 3, 4],
  "model": "Polo",
  "year": 1991
},{
  "brand": "Ford",
        "l": ["Aaron Beta", 2, 3, 4],
  "model": "Mustang",
  "year": 1964
}]

with open("1.xml", "w") as xml:
    print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>", file=xml)
    print("<garage>", file=xml)
    for car in cars:
        vorname = car["l"][0].split(" ")[0]
        nachname = car["l"][0].split(" ")[1]
        
        print(vorname, nachname)
        if car["year"] > 1980:
            print("\t<car>", file=xml)
            for key in car.keys():
                print("\t\t"+"<"+str(key)+">"+str(car[key])+"<\\"+str(key)+">", file=xml)
            print("\t<\\car>", file=xml)
    print("<\\garage>\n", file=xml)
```

# 2: From JSON to XML

```python
import json

with open("bibsonomy.json", "r") as file:
    data = json.load(file)
```

```python
type(data["items"])
```

```python
data["items"]
```

```python
edited_items = []
# Auswahl der Informationen, die ausgewählt werden sollen:
desired_entries = ['type', 'year', 'publisher', 'author', 'label', 'url']

# für jedes element im eingelesenen JSON file (über data.keys() lässt sich leicht herausfinden, wo die relevanten Daten stecken):
for item in data["items"]:
    # erstelle ein neues leeres dictionary für die Daten des Elements
    item_export = {}
    # für jeden gewünschten Eintrag
    
    # hier können Filterkriterien hinzugefügt werden (hier z.B. erschienen nach 2009 und mit weniger als 4 Autoren):
    if int(item["year"])>2009  and len(item["author"])<4:
        for entry in desired_entries:
            # ... überprüfe, ob es einen Eintrag gibt (um Fehler zu vermeiden)
            if entry in item:
                # füge den Eintrag den Daten zum Element hinzu
                item_export[entry] = item[entry]
        # am Ende: füge den vollständigen (gefilterten) Eintrag der zu exportierenden Liste hinzu
        print(edited_items)
        edited_items.append(item_export)
```

## String Formatting

```python
# String formatting

cars2 = [
    {
  "brand": "Peugeot",
  "model": "206",
  "year": 1985
},{
  "brand": "VW",
  "model": "Polo",
  "year": 1991,
  "color": "red"
},{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}]

txt="I am driving my {} {} {} from {} to work."
print(txt.format(cars2[1]["color"], cars2[1]["brand"], cars2[1]["model"], cars2[1]["year"]))
```

```python
# alternative:
values = list(cars2[1].values())
print(values)

# Asteriks "entpackt" liste
print(*values)

# deswegen funktioniert das
print("I am driving my {3} {0} {1} from {2} to work.".format(*(values)))

# wie das:
print("I am driving my {3} {0} {1} from {2} to work.".format(values[0], values[1], values[2], values[3]))
```
